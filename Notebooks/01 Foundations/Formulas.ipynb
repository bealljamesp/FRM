## 1. Probability & Statistics

### Moments of a Distribution

- **Mean (1st Moment):** $E[X] = \mu$
    
- **Variance (2nd Moment):** $\sigma^2 = E[(X - \mu)^2]$
    
- **Skewness (3rd Moment):** $S = E[\frac{(X - \mu)^3}{\sigma^3}]$ (Measures asymmetry; Negative Skew = Left Tail risk)
    
- **Kurtosis (4th Moment):** $K = E[\frac{(X - \mu)^4}{\sigma^4}]$ (Normal distribution $K=3$; Leptokurtic $K > 3$)
    

### Portfolio Math (Linear Algebra Framework)

- **Portfolio Variance (2 Assets):** $\sigma_p^2 = w_1^2\sigma_1^2 + w_2^2\sigma_2^2 + 2w_1w_2\sigma_1\sigma_2\rho_{12}$
    
- **Correlation:** $\rho_{12} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$

## 2. Value-at-Risk (VaR)

VaR represents the maximum loss over a target horizon such that there is a low probability (significance level $\alpha$) that the actual loss will be larger.

### Parametric VaR (Normal Distribution Assumption)

- **Formula:** $VaR_{\alpha} = \mu - (z_{\alpha} \times \sigma)$
    
- **95% Z-Score:** $-1.645$
    
- **99% Z-Score:** $-2.326$
    

### Time & Volatility Scaling (Square Root of Time Rule)

- **$T$-day VaR:** $VaR_{1\text{-day}} \times \sqrt{T}$
    
- **$T$-day Volatility:** $\sigma_{1\text{-day}} \times \sqrt{T}$

## 3. Fixed Income & Valuation

### Bond Pricing

- **Price:** $P = \sum_{t=1}^{T} \frac{C}{(1+y)^t} + \frac{F}{(1+y)^T}$
    
- **Macaulay Duration:** The weighted average time to receive cash flows.
    
- **Modified Duration ($D^*$):** Measures price sensitivity to yield changes.
    
    - $D^* = \frac{D_{Mac}}{1 + \frac{y}{k}}$ (where $k$ is compounding frequency)
        
    - $\Delta P \approx -P \times D^* \times \Delta y$

## 4. Derivatives Pricing (BSM)

### Black-Scholes-Merton Call Price

$C = S_0 N(d_1) - Ke^{-rT} N(d_2)$

- $d_1 = \frac{\ln(S_0/K) + (r + \sigma^2/2)T}{\sigma\sqrt{T}}$
    
- $d_2 = d_1 - \sigma\sqrt{T}$

## 5. Risk Management Heuristics

- **Jensen's Inequality:** $E[f(X)] \geq f(E[X])$ for convex functions (crucial for option pricing).
    
- **Central Limit Theorem:** Distribution of sample means approaches $N(\mu, \sigma^2/n)$ as $n$ increases, regardless of the underlying distribution shape.
